```python
import pandas as pd
df = pd.read_csv('brcancer.csv')
```

```python
data= df.copy()
```

```python
data.head(3)
```

```python
data['surgery'] = data.apply(lambda row: 'yes' if row['horTh'] == 'no' and row['tgrade'] >= 'II' else 'no', axis=1)
```

```python
data['radTh'] = data.apply(lambda row: 'yes' if row['surgery'] == 'yes' and row['tsize'] >= 50 else 'no', axis=1)
```

```python
data.head(6)
```

```python
data.isnull().sum()
```

```python
data.to_csv('processed_data.csv', index=False)
```

```python
from sklearn.preprocessing import LabelEncoder

# Assuming 'df' is your DataFrame and 'column_name' is the column containing categorical values
# Replace 'df' and 'column_name' with the actual names of your DataFrame and column

# Initialize LabelEncoder
label_encoder = LabelEncoder()

# Fit and transform the column with LabelEncoder
data['horTh'] = label_encoder.fit_transform(data['horTh'])
data['surgery'] = label_encoder.fit_transform(data['surgery'])
data['radTh'] = label_encoder.fit_transform(data['radTh'])
# data['menostat'] = label_encoder.fit_transform(data['menostat'])
# data['tgrade'] = label_encoder.fit_transform(data['tgrade'])


# Create a legend
legend = dict(zip(label_encoder.classes_, label_encoder.transform(label_encoder.classes_)))

print("Legend:")
for key, value in legend.items():
    print(f"{key}: {value}")


value_counts = data['menostat'].value_counts()
value_counts2 = data['tgrade'].value_counts()


print("Count of each categorical value:")
print(value_counts)
print(value_counts2)

data.head(8)
```
